```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import ast

filepath = r"C:\Users\jjap2\Poleto SCE\Poleto-SCE\dond_game_data.csv"
df = pd.read_csv(filepath)
```

```python
# This cell excludes postgame offers
# Needed for when creating banker thanks to the articles relevant to functions of the banker in deal or no deal
# Should result in a count of 730, otherwise something is wrong
df = df[df["Postgame"] != 1]
```

```python
# This cell excludes when offers = 0
# AKA when contestants stick with their cases and don't switch
# count = 717
df = df[df["Offer"] != 0]
```

```python
# This cell splits remaining values into a string, unsure if this is needed yet
# RUN ONLY IF NEEDED
# Convert "Remaining Values" from string to list
df["Remaining Values"] = df["Remaining Values"].apply(ast.literal_eval)

# Extract unique values
all_values = set()
for values in df["Remaining Values"]:
    all_values.update(values)

for value in sorted(all_values):
    df[f"Case_{value}"] = df["Remaining Values"].apply(lambda x: 1 if value in x else 0)
```
